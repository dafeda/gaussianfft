```python
import gaussianfft
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
nx, ny = 100, 100
dx, dy = 1, 1
v = gaussianfft.variogram('general_exponential', 12, 12, power=1.5, azimuth=0)
```

```python
corr = [v.corr(x) for x in np.linspace(0, nx * dx, 100)]
plt.plot(corr)
```

```python
gaussianfft.seed(423)
f = gaussianfft.advanced.simulate(v, nx, dx, ny, dy, padx=20, pady=20)
f_arr = np.array(f).reshape((nx, ny), order='F')
plt.imshow(f_arr, interpolation='None')
plt.colorbar()
```

```python
gaussianfft.seed(42)
f = gaussianfft.advanced.simulate(v, nx, dx, ny, dy, padx=100, pady=100)
f_arr = np.array(f).reshape((nx, ny), order='F')
plt.imshow(f_arr, interpolation='None')
plt.colorbar()
```
