```python
import nrlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [15.0, 10.0]
```

```python
# Constants throughout this notebook
nx, ny, nz = 250, 250, 50
dx, dy, dz = 10, 10, 10
lx = np.arange(0, nx) * dx
ly = np.arange(0, ny) * dy
lz = np.arange(0, nz) * dz
# ---

def sim(a, d):
    v = nrlib.variogram('spherical', 1000.0, 500.0, 100, azimuth=a, dip=d)
    field = nrlib.simulate(v, nx, dx, ny, dy, nz, dz)
    field_array = np.array(field).reshape((nx, ny, nz), order='F')
    return field_array
```

```python
def plotx(f):
    plt.contourf(ly,
                 lz,
                 f[int(nx/2), :, :].T,
                 40)
    
def ploty(f):
    plt.contourf(lx,
                 lz,
                 f[:, int(ny/2), :].T,
                 40)
    
def plotz(f):
    plt.contourf(lx,
                 ly,
                 f[:, :, int(nz/2)].T,
                 40)
```

# Azimuth = 0, Dip = 0

```python
f = sim(0, 0)
plt.rcParams['figure.figsize'] = [15, 15]
plt.subplot(311)
plotx(f)
plt.gca().set_aspect('equal')
plt.title('x-slice')
plt.subplot(312)
ploty(f)
plt.gca().set_aspect('equal')
plt.title('y-slice')
plt.subplot(313)
plotz(f)
plt.gca().set_aspect('equal')
plt.title('z-slice')
```
