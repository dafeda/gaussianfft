```python
import nrlib
import numpy as np
import matplotlib.pyplot as plt


v = nrlib.variogram('general_exponential', 2000.0, 1000.0, azimuth=30.0)
nx, ny = 100, 125
dx, dy = 20.0, 20.0
nrlib.seed(42)
f_vector = nrlib.simulate(v, nx, dx, ny, dy)
f = np.array(f_vector).reshape((nx, ny), order='F')

x = np.arange(0, nx) * dx
y = np.arange(0, ny) * dy
plt.contourf(y, x, f, 30)
plt.colorbar()
plt.xlabel('East')
plt.ylabel('North')
plt.show()
```
